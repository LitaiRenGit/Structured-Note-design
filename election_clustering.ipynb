```python
def minute_to_intraday(series,f_list):
    """
    series: a pd.Series with datetime index
    f_list: a list of function, each of which with input as pd.series/np.array
    return: a dataframe whose columns are calculated result of f_list
    """
    return series.groupby(series.index.date).agg(f_list)

import numpy as np
def Range(series): return series.max()-series.min()
def first(series): return series[0]
def last(series): return series[-1]
minute_to_intraday(Log_returns_1,[np.mean, Range, np.std, len, first, last])# example for minute_to_intraday
```

```python
def minute_to_interday(series,f):
    import numpy as np
    def first(series): return series[0]
    def last(series): return series[-1]
    temp=minute_to_intraday(series,[first,last])
    return f(temp['first'],temp['last'].shift(periods=1))

minute_to_interday(Log_returns_1,lambda x,y:x-y)
```
